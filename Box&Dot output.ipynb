```python
#get distance matrix - euclidean distance
def get_dist_btw_pts(x):
    return distance_matrix(x, x, p=2)

def get_boxplot_timeMatrix(points):
    dist_matrix = get_dist_btw_pts(points)
    time_matrix = dist_matrix*60 #convert to mins
    return time_matrix.tolist()
```
